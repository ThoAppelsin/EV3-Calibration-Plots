```julia
using GLM, DataFrames, Plots, Distributions, HypothesisTests, StatPlots, Measures

pyplot()
theme(:wong)

slicedims(mat, dim) = broadcast(x -> selectdim(mat, dim, x), 1:size(mat, dim))
ivcat(mat) = slicedims(mat, 1)
```

# Reading the Data

```julia
fn_tabs(str) = count(x -> x == '\t', str)

file = readchomp("data.txt");
lines = split(file, "\n");
filter!(l -> length(strip(l)) > 0, lines);
indent_lines = map(l -> (fn_tabs(l), strip(l)), lines);

function hierarch(indent_lines)
    inds = findall(i_l -> i_l[1] == indent_lines[1][1], indent_lines)
    ends = [inds[2:end]; length(indent_lines) + 1]

    if length(inds) == length(indent_lines)
        return map(i_l -> map(x -> parse(Float64, x), split(i_l[2])), indent_lines)
    end

    d = Dict()
    for (i, e) in zip(inds, ends)
        d[indent_lines[i][2]] = hierarch(indent_lines[i+1:e-1])
    end

    return d
end

data = hierarch(indent_lines)
```

# Sensor Calibration Plots 

```julia
function linefit(data, title)
    ols = lm(@formula(Sensor ~ Actual), data)
    intercept, slope = coef(ols)
    
    scatter(data[:Actual], data[:Sensor], title=title,
        xlabel="Actual", ylabel="Sensor", label="Data", m=(0.2, :x, 7))

    plot!(x -> intercept + slope * x, [minimum(data[:Actual]), maximum(data[:Actual])],
        label="Fit: y = $(round(intercept, digits=3)) +" *
        "$(round(slope, digits=3))⋅x")
end
```

# Ultrasonic Sensor

```julia
ultrasonic_data = DataFrame(
    vcat(([row[1] measured] for row in data["Ultrasonic"] for measured in row[2:end])...), [:Actual, :Sensor])

linefit(ultrasonic_data, "Ultrasonic")
```

# Light - Color Sensor

```julia
function barfit(data, title)
#     xticks = map(x -> String(x)[2:end], names(data))
    σy, μy, ny = ivcat(colwise([x -> -(confint(OneSampleTTest(x))...) / -2, mean, length], data))
    plot(["Red", "Green", "Blue", "Intensity"], μy, t=:yerror, yerror = σy, title = title, ylims = (0, 70))
end

function analyzecolor(colordata, lights, papercolor)
    darklight = lights ? "Light" : "Dark"
    data = DataFrame(hcat(colordata[darklight][papercolor]...)' .* [1 1 1 100], [:Red, :Green, :Blue, :Intensity])
    barfit(data, papercolor * (lights ? " under Light" : " in the Dark"))
end

function analyzecolor(lights)
    darklight = lights ? "Light" : "Dark"
    plot((analyzecolor(data["Color"], lights, papercolor) for papercolor in keys(data["Color"][darklight]))...,
        xtickfont=font(10, "Verdana", colorant"gray40"),
        titlefont=font("Verdana", colorant"gray20"),
        margin=5mm,
        size=(900,600))
end
```

## Light

```julia
plot((analyzecolor(data["Color"], true, papercolor) for papercolor in keys(data["Color"]["Light"]))...,
    xtickfont=font(10, "Verdana", colorant"gray40"),
    titlefont=font("Verdana", colorant"gray20"),
    margin=5mm,
    size=(900,600))
```

## Dark

```julia
plot((analyzecolor(data["Color"], true, papercolor) for papercolor in keys(data["Color"]["Dark"]))...,
    xtickfont=font(10, "Verdana", colorant"gray40"),
    titlefont=font("Verdana", colorant"gray20"),
    margin=5mm,
    size=(900,600))
```
